```python
# functions to check all possibilites of trump in twitter text

import re
#for e.g if text has realdonald, will return donald
def find_president(text):
    matches = re.findall(r'trump|donald|president', text, re.IGNORECASE)
    if matches:
        return ','.join(matches)
#returns the whole word for e.g if text has realdonald, will return realdonald
def find_president_word(text):
    matches = re.findall(r'\s*[^\s]*trump|donald|president[^\s]*\s*', text, re.IGNORECASE)
    if matches:
        return ','.join(matches)
```

```python
df['trump'] = df.text.apply(find_president)
df['trumpword'] = df.text.apply(find_president)
```

```python
# function to find possible emotions highlighted in text
def find_emotions(text1):
    matches = re.findall(r'good|very good|nice', text1, re.IGNORECASE)
    if matches:
        return ','.join(matches)
```

```python
df['Emotions'] = df.text.apply(find_emotions)
```
